```python
# add a ones column - this makes the matrix multiplication work out easier
data.insert(0, 'Ones', 1)

# set X (training data) and y (target variable)
cols = data.shape[1]
X = data.iloc[:,0:cols-1]
y = data.iloc[:,cols-1:cols]

# convert to numpy arrays and initalize the parameter array theta
X = np.array(X.values)
y = np.array(y.values)
theta = np.zeros(3)
```

Sigmoid Function

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
a = np.array(list(range(-10, 10)))
```

```python
plt.figure(figsize=(12, 8))
plt.plot(a, sigmoid(a))
```

sigmoid works fine

Cost funcion

```python
def cost(theta, X, y):
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    
    term1 = np.multiply(-y, np.log(sigmoid(X*theta.T)))
    term2 = np.multiply(1-y, np.log(1-sigmoid(X*theta.T)))
    calc = np.sum(term1 - term2)
    return calc/len(X)
```

```python
a = cost(theta, X, y)
a
```

Cost function works fine

```python
def gradient(theta, X, y):
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    grad = np.zeros(X.shape[1])
    
    error = sigmoid(X * theta.T) - y
    for i in range(0, X.shape[1]):
        term = np.multiply(error, X[:,i])/len(X)
        grad[i] = np.sum(term)
    return grad
```

```python
gradient(theta, X, y)
```

Gradient works fine

```python
import scipy.optimize as opt
result  = opt.fmin_tnc(func=cost, x0=theta, fprime=gradient, args=(X, y))
result
```

```python
new_theta = result[0]
```

```python
def predict(X, theta):
    theta = np.matrix(theta)
    X = np.matrix(X)
    prediction = sigmoid(X * theta.T) > .5
    return np.array(prediction)
```

```python
prediction = predict(X, new_theta)
```

```python
temp  = y == 1
```

```python
count = 0
for i in range(0, len(prediction)):
    if prediction[i] == temp[i]:
        count = count + 1
```

```python
accuracy = (count*100)/len(X)
print("The accuracy is :",count)
```
