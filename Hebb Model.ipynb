```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
T = np.array([0, 0, 0, 1])
```

```python
def addBias(X):
    return np.insert(X, 0, 1, axis=1)
```

```python
def binaryToBipolar(X):
    X[X == 0] = -1
    return X
```

```python
def getError(X, T, W):
    tmp = (W*X).sum(axis=1)
    tmp[tmp >= 0] = 1
    tmp[tmp < 0] = -1
    tmp = (tmp != T).sum()
    return tmp
```

```python
def train(X, T, W):
    X = binaryToBipolar(X)
    T = binaryToBipolar(T)
    
    error = []
    error.append(getError(X, T, W))
    
    for i in range(0, len(X)):
        W = W + X[i]*T[i]
        error.append(getError(X, T, W))
        
    return W, error
```

Training

```python
X = addBias(X)
W = np.zeros(X.shape[1])
W , error = train(X, T, W)
W
```

```python
plt.figure(figsize=(12,8))
plt.plot(error)
```
